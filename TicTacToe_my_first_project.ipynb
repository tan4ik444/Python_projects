```python
player = ''
player1 = ''
player2 = ''
def x_o_choice(): #to ask if player want to play with X or O
    global player1
    global player
    while str.upper(player1) not in ('X', 'O'):
        player1 = input ('Player 1: Do you want to be X or O? ')
    else:
        print('Player 1 will go first')
    player=player1

def define_player2(): #to define marker for player 2
    global player2
    if str.upper(player1)=="X":
        player2="O"
    else:
        player2='X'
    return player2

def number_board(): #function that can print out a board with numbers of cells
    print('Please use the number of cell for your step\n| 1 | 2 | 3 |\n--------------\n| 4 | 5 | 6 |\n--------------\n| 7 | 8 | 9 |\n\n')
#board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
def display_board(): #function that can print out a board
    global board
    from IPython.display import clear_output
    clear_output()
    number_board()
    print ('| {} | {} | {} |\n--------------\n| {} | {} | {} |\n--------------\n| {} | {} | {} |\n' .format (board[0],board[1],board[2],board[3],board[4],board[5],board[6],board[7],board[8]))
def player_input(): #function that can take in a player input and assign their marker as 'X' or 'O'. Think about using while loops to continually ask until you get a correct answer.
    p_input=0
    while p_input not in range (1,10):
        p_input = int(input ('Make your choice: 1-9'))
    else:
        return p_input

def place_marker(board, player, p_input): #function that takes in the board list object, a marker ('X' or 'O'), and a desired position (number 1-9) and assigns it to the board.
    board[p_input-1]=str.upper(player)
    from IPython.display import clear_output
    clear_output()
    display_board()

def win_check(mark): #function that takes in a board and a mark (X or O) and then checks to see if that mark has won
    global board
    if board[0]==board[1]==board[2]==mark or board[3]==board[4]==board[5]==mark or board[6]==board[7]==board[8]==mark or board[0]==board[3]==board[6]==mark or board[1]==board[4]==board[7]==mark or board[2]==board[5]==board[8]==mark or board[0]==board[4]==board[8]==mark or board[2]==board[4]==board[6]==mark:
        return True
    else:
        return False
    
def full_board_check(): #Write a function that checks if the board is full and returns a boolean value. True if full, False otherwise
    global board
    if ' ' not in board:
        return True
    else:
        return False
    
def play_choice():   #function that asks for a player's next position (as a number 1-9) and then uses the function space_check() to check if it's a free position
    global board
    p_input=0
    while p_input not in range (1,10):
        p_input = int(input ('Make your choice of free cell: 1-9'))
        if board[p_input-1]==' ':
            return p_input
        else:
            while p_input<10 and board[p_input-1]!=' ':
                p_input = int(input ('This cell is not free, please try again: 1-9'))
                from IPython.display import clear_output
                clear_output()
                display_board()
            return p_input        
def replay(): #function that asks for game start or exit
    r_answer=''
    while str.upper(r_answer) not in ('YES', 'NO'):
        r_answer = input ('Are you ready to play? Yes/No')
    if str.upper(r_answer )=='NO':
        print ('Good bye!')
        return False
    else:
        return True

print('Welcome to the Tic Tac Toe game!')
player1 = ''
player2 = ''
while True:
    x_o_choice()
    define_player2()
    while replay():
        board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
        display_board()
        pass
        place_marker(board, player, int(player_input()))
        while not win_check('X') and not win_check('O'):
            while not full_board_check():
                print("player 2 turn:")
                player=player2
                place_marker(board, player, int(play_choice()))
                if win_check('X') or win_check('O'):
                    break
                print("player 1 turn:")
                player=player1 
                place_marker(board, player, int(play_choice()))
                if win_check('X') or win_check('O'):
                    break
            if win_check(player1):
                    print('Congratulations, {}!'.format(player1))
                    break
            elif win_check(player2):
                    print('Congratulations, {}!'.format(player2))
                    break
            elif full_board_check():
                break
    if not replay():
        break
```
